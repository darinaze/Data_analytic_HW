1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 10*x - 5
```

```python
def computation_derivative_f(f, x, dx=1e-6):
    return (f(x + dx) - f(x - dx)) / (2 * dx)
```

```python
computation_derivative_f(f,16)
```

```python
x = np.linspace(0, 10, 50)
y = 10*x - 5

plt.plot(x, y, label='f(x) = 10x - 5', color='blue', linewidth=2)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік функції f(x) = 10x - 5')
plt.legend()
plt.grid()

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import math
```

```python
def f_sin(x):
    return math.sin(x)
```

```python
def f_exp(x):
    return 1/x if x != 0 else float('inf')
```

```python
def derivative_f(f, x, dx=1e-6):
    return (f(x + dx) - f(x - dx)) / (2 * dx)
```

```python
print("sin(x) ", derivative_f(f_sin, math.pi / 4))
print("exp ", derivative_f(f_exp, 8))
```

```python
import sympy as sp
x = sp.symbols('x')
sin_x = sp.sin(x)

derivative_f_sin = sp.diff(sin_x, x)
print(derivative_f_sin)
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y, label='f(x) = sin(x)', color='blue', linewidth=2)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік функції f(x) = sin(x)')
plt.legend()
plt.grid()

plt.show()
```

```python
x = sp.symbols('x')
exp_y = sp.exp(x)

derivative_f_exp = sp.diff(exp_y, x)
print(derivative_f_exp)
```

```python
x = np.linspace(0, 10, 50)
y = np.exp(x)

plt.plot(x, y, label='f(x) = exp(x)', color='blue', linewidth=2)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік функції f(x) = exp(x)')
plt.legend()
plt.grid()

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
derivative = sp.diff(f, x)
print('похідна :' , derivative)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 5*x + 1

def f_derivative(x):
    return 2*x - 5

x = np.linspace(0, 100, 200)
y_f = f(x)
y_f_derivative = f_derivative(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_f, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x, y_f_derivative, label="f'(x) = 2x - 5", color='green')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Функція та її похідна")
plt.grid(True)

plt.show()
```
